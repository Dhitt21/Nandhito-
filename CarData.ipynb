```python
# Berikut kode untuk memperoleh rata-rata standar deviasi, nilai ekstrem.
numericdframe = ["Year","Selling_Price","Present_Price", "Kms_Driven"]
for i in range(len(numericdframe)):
    data = numericdframe[i]
    print(str(i+1) + ". " + data)
    print("Rata-rata: " + str(dframe[data].mean()))
    print("Standar Deviasi: "+ str(dframe[data].std()))
    print("Nilai ekstrem: ")
    print("\tMax: " + str(dframe[data].max()))
    print("\tMin: " + str(dframe[data].min()))
    print()
```
